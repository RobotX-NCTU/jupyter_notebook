# Feature Matching

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('tutorial_data/box.png',0)          # queryImage
img2 = cv2.imread('tutorial_data/box_in_scene.png',0) # trainImage
```

get SIFT features

```python
sift = cv2.xfeatures2d.SIFT_create()
(kps1, descriptors1) = sift.detectAndCompute(img1,None)
(kps2, descriptors2) = sift.detectAndCompute(img2,None)
```

brute force matching

```python
bf = cv2.BFMatcher()
matches = bf.knnMatch(descriptors1,descriptors2, k=2)
```

calculate good matches

```python
good = []
for m,n in matches:
    if m.distance < 0.75*n.distance:
        good.append([m])
```

```python
outImg = 0
img3 = cv2.drawMatchesKnn(img1,kps1,img2,kps2,good,outImg,flags=2)

plt.imshow(img3),plt.show()
```

FLANN matching

```python
# FLANN parameters
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks=50)   # or pass empty dictionary

flann = cv2.FlannBasedMatcher(index_params,search_params)

matches = flann.knnMatch(descriptors1,descriptors2,k=2)
```

```python
# Need to draw only good matches, so create a mask
matchesMask = [[0,0] for i in xrange(len(matches))]

# ratio test as per Lowe's paper
for i,(m,n) in enumerate(matches):
    if m.distance < 0.7*n.distance:
        matchesMask[i]=[1,0]
```

```python
draw_params = dict(matchColor = (0,255,0),
                   singlePointColor = (255,0,0),
                   matchesMask = matchesMask,
                   flags = 0)

img3 = cv2.drawMatchesKnn(img1,kps1,img2,kps2,matches,None,**draw_params)

plt.imshow(img3),plt.show()
```
